## Step 2: Clean data

##### Do some rudimentary cleaning. Rename any columns that are poorly named, shorten any strings that are too long, and check for missing values (replace them and explain rationale if it makes sense).

```python
#Creating a new dataset that I will clean in-place
billboard_clean = billboard.copy()
```

```python
#Dropping useless variables: "year" is all the same, so it adds nothing; dropping all columns with all NaNs
billboard_clean.dropna(how='all', axis=1, inplace=True)
billboard_clean.drop('year', axis=1, inplace=True)
billboard_clean.head() #checking if correct
```

```python
#Renaming some variables
#first, rename 'artist.inverted" to be simpler and 'time' to 'length' to be clearer.
billboard_clean.rename(columns = {'artist.inverted' : 'artist', 'time' : 'length'}, inplace = True)

#next, getting rid of superfluous characters
billboard_clean.columns = [col.strip('x') for col in billboard_clean.columns]

#converting decimals to underscores to make dot-notation possible with all columns
billboard_clean.columns = [col.replace('.', '_') for col in billboard_clean.columns]

billboard_clean.head() #tests if it all worked
```

```python
# Converting 1st_week to float for consistency with the other Nth.week columns (NaN values are not 
# supported in an integer dtype) so I can do things with them together and do the same things to them.
billboard_clean.loc[:, '1st_week'] = billboard_clean.loc[:, '1st_week'].astype(float)
billboard_clean.dtypes['1st_week'] #testing if it worked
```

```python
#Now I am going to convert some datatypes to something more useful
#convert dates form objects to datetimes
billboard_clean['date_entered'] = pd.to_datetime(billboard_clean['date_entered'], format='%Y-%m-%d')
billboard_clean['date_peaked'] = pd.to_datetime(billboard_clean['date_peaked'], format='%Y-%m-%d')

#check if it appeared to do it right
print(billboard_clean['date_entered'].head(), billboard['date.entered'].head())
print(billboard_clean['date_peaked'].head(), billboard['date.peaked'].head())

#test that it is now the correct dtype
billboard_clean[['date_entered', 'date_peaked']].dtypes
```

```python
# 'length' column is listed as a object-dtype but would be more useful as a float.
# 1st made a function to convert a string in format 'min:sec' to a float in minutes
def minsec_to_float(string):
    min_sec = string.split(':')
    min_float = float(min_sec[0]) + float(min_sec[1])/60
    return min_float
# Test the function
minsec_to_float('3:50')
```

```python
#apply the function to 'length', return to new column in case of errors/bugs/fails
billboard_clean['length'] = billboard_clean['length'].apply(minsec_to_float)
#alternatively:
# billboard_clean['length'] = billboard_clean['length'].apply(lambda x: float(string.split(':')[0]) + float(string.split(':')[1])/60)

# #print the new and old columns side-by-side to see if it seems to have worked
#pd.DataFrame([billboard_clean['length'], billboard['time']]).head()

#JB an approach to better format:
pd.DataFrame({'length':billboard_clean['length'], 'time':billboard['time']}).head()
```

```python
# investigate 'time_float'
print(billboard_clean['length'].describe())
```

```python
#Now I want to create some useful calculated variables: Weeks on Chart, Highest Ranking, Weeks to Peak:
billboard_clean.ix[:, 'weeks_on_chart'] = billboard_clean.ix[:, '1st_week':'65th_week'].notnull().sum(axis=1)
billboard_clean.ix[:, 'best_ranking'] = billboard_clean.ix[:, '1st_week':'65th_week'].min(axis=1)

days_to_peak = billboard_clean['date_peaked'] - billboard_clean['date_entered']
billboard_clean['weeks_to_peak'] = days_to_peak.apply(lambda l: l.days)/7
billboard_clean.head() #test if it worked
```

```python
#function to take stats.describe() and clean it up a bit and add a median
from scipy import stats
def summarize(column):
    d = stats.describe(column)
    print "no. values: {}".format(d[0])
    print "min: {:.3}, max: {:.3}".format(d[1][0]/1.0, d[1][1]/1.0)
    print "mean: {:.3}".format(d[2]/1.0)
    print "median: {:.3}".format(np.median(column)/1.0)
    print "st dev: {:.3}".format(d[3]**.5)
    print "skew: {:.3}".format(d[4]/1.0)
    print "kurt: {:.3}".format(d[5]/1.0)
    #had to divide by 1.0 in order to ensure they would always be floats; otherwise the formatting raises an error
```

# Weeks on the Billboard Top 100

```python
#summarizing and visualizing spread of weeks a song spends on the chart
summarize(billboard_clean.weeks_on_chart)
sns.distplot(billboard_clean.weeks_on_chart, hist_kws={'alpha':0.2})
plt.title('Distribution of How Long Songs Spend on the Charts\n', size=15)
```

There's a HUGE peak around 20, with the rest appearing somewhat like a right-skewed distribution. The mean is 16.7, which appears to be skewed heavily by the large number of songs reaching around 20 weeks.

```python
#investigating the peak at 20 a bit further
billboard_clean.weeks_on_chart.value_counts().head()
```

A large chunk of songs make it *exactly* 20 weeks on the charts. This seems like too strong a pattern to be natural, so I did more investigation. According to  https://en.wikipedia.org/wiki/Billboard_Hot_100, the Billboard Hot 100 charts have a rule regarding "recurrent" songs (songs that have been on the charts a long time) in order to keep the charts current. While I can't find data on what the exact rule was in 2000, current rule is that a song is moved to "recurrent" status (and is therefore off the Billboard Hot 100 chart) if it "has spent 20 weeks on the Hot 100 and fallen below position number 50" or "ranking below number 25 after 52 weeks". While exceptions are made, it's clear that this has a major impact on the data, and songs that last exactly 20 weeks have likely not have fallen off the charts "organically". In addition, songs that fall below 50 after peaking for a long time may still be in the top 100 but are not on the charts.

```python
# To confirm if these data follow what I read on wikipedia, I charted the distribution of all songs below
sns.set_style('white')
weeks = billboard_clean.ix[:, '1st_week':'65th_week'].transpose()
weeks.index = range(1, 66)
fig1 = plt.figure()
for col in weeks.columns:
    if min(weeks[col]):
        plt.plot(weeks[col], color='b', alpha=.15) #made them a bit see-through so you can see some density
plt.gca().invert_yaxis() #axis inverted because 1 is "higher" than 100 in this dataset.
plt.xlabel('weeks since entered chart')
plt.ylabel('ranking (axis inverted)')
plt.title('Each Song\'s Trajectory after they First Enter the Charts\n', size=15)
```

```python
mpld3.save_html(fig1, 'fig1.html')
```

This chart shows that, with only 2-3 exceptions, no songs last over 20 weeks unless they were above about 50 on the charts. This is consistent with the evidence found on Wikipedia.

```python
# I'd like to examine the distribution of songs that didn't reach exactly 20 weeks - Songs that lasted fewer
# than 20 weeks would have fallen off the charts naturally, and the ones that lasted longer, but not exactly 20 weeks,
# would have fallen off the charts slightly more naturally, although they would have been removed early as well, on
# their way 'down'
summarize(billboard_clean.weeks_on_chart[billboard_clean.weeks_on_chart != 20])
sns.distplot(billboard_clean.weeks_on_chart[billboard_clean.weeks_on_chart != 20], hist_kws={'alpha':0.2})
plt.title('Distribution of Songs\' Best Ranking\nwithout songs that last exactly 20 weeks\n', size=15)
```

This distribution is much more normal, and the kurtosis and right skew is to be expected because:
- songs that *would* have gone above 20 weeks but were booted off the chart would have filled the right side out a little.
- Certain songs get very popular and stay popular a very long time, but they are relatively few and far between, so you'd expect for some of them to appear and thus skew the data
- Songs can't last fewer than 0 weeks on the charts, creating a 'floor' effect. 

This distribution is more normal and therefore parametric tests would be more robust to this distribution's violations of the normality assumptions than would the original distribution, especially if the skew is in the same direction for subsets of these data. It would take more analysis to determine if it is better to leave off this data or keep it.

# Weeks to Peak

```python
#summary and visualization of weeks_to_peak
summarize(billboard_clean['weeks_to_peak'])
sns.distplot(billboard_clean['weeks_to_peak'], hist_kws={'alpha': 0.2})
plt.title('Distribution of How Long it Takes Songs to Peak on the Charts\n', size=15)
```

There's a good chunk who peak soon after they enter the charts, but there's also a distribution around about 7 with a right skew. Very few songs take more than 20 weeks to peak, which may be one reason Billboard decided to drop songs after they've been on the charts for that long.

```python
#investigating the relationship between weeks-to-peak and length on the charts:
sns.regplot(billboard_clean.weeks_to_peak, billboard_clean.weeks_on_chart)
print "r: {:.2}".format(stats.pearsonr(billboard_clean.weeks_to_peak, billboard_clean.weeks_on_chart)[0])
plt.title('Weeks to Peak vs. Weeks on Chart\n', size=15)
```

There appears to be a very strong relationship between how long it takes a song to peak and how long it lasts on the charts - songs that peak early don't last as long - so if they're on the downward swing early on they're not likely to make a comeback. The Pearson's 'r' correlation coefficient is 0.86, which is quite high, indicating a strong positive relationship.

```python
#investigating the relationship between weeks-to-peak and length on the charts:
sns.regplot(billboard_clean.weeks_to_peak, billboard_clean.best_ranking)
plt.gca().invert_yaxis()
print "r: {:.2}".format(stats.pearsonr(billboard_clean.weeks_to_peak, billboard_clean.best_ranking)[0])
plt.title('Weeks to Peak vs. Songs\' Highest Rank\n', size=15)
```

There also appears to be a relationship between how long it takes a song to peak and high high it gets on the charts: songs that reach a higher ranking tend to take longer to peak. The relationship doesn't appear as strong. The Pearson's 'r' value suggests the same thing since since it's -0.6 (negative because the desired direction), which is a weaker correlation. The correlation may be influenced by the number of songs that hit on or near #1 even though they took a long time to hit their peaks.

### How long it takes a song to peak is highly correlated to how well it will do overall on the Charts, especially in terms of staying power. Songs that may take a while to peak actually end up lasting longer, so don't discount a song for climbing slowly.

Considering the strong relationship between how quickly a song peaks and how long it will stay on the charts, I was surprised that the the 1st week's ranking was such a weak predictor of length on the charts. I wanted to see if subsequent weeks were more predictive, so I decided to do some further exploration of the relationship between ranking on subsequent weeks and length of time on the charts.

```python
for col in billboard_clean.ix[:, '1st_week':'10th_week']:
    print "r: {:.2}".format(stats.pearsonr(billboard_clean[col][billboard_clean[col].isnull() == False], billboard_clean.weeks_on_chart[billboard_clean[col].isnull() == False])[0])
    sns.regplot(billboard_clean[col], billboard_clean.weeks_on_chart)
    plt.gca().invert_xaxis()
    plt.title('{} Performance vs. Weeks on Chart\n'.format(col), size=15)
    plt.show()
```

As I suspected, the predictive power of any given week's ranking increases week after week. Of course, some of that is an artifact of the data because, for instance, if a song has made it to the 10th week then it has to have lasted at least 10 weeks on the charts. # Track Names

```python
#I have to first count the frequency of each word using collections.
from collections import Counter 
import re
all_names = ' '.join(list(billboard['track'])) #will create a string with all rows concatenated together by a space 
#needed 'list()' because it is a pandas series otherwise, which .join() won't work on
all_names_simplified = re.sub("[^A-Za-z0-9 ]", '', all_names.lower()) #lowercases everything and removes special characters so it will count accurately
all_names_list = all_names_simplified.split(' ') #will split everything into its separate words
counter_items = Counter(all_names_list) #creates a dictionary of words:wordcount
df_count = pd.DataFrame(counter_items.items(), columns=['a', 'b']) #converts to dataframe so I can sort (used .items() to get rid of error)
df_count = df_count.sort_values('b', ascending=False) #sorts ascending by word count ('b')

#i want to get rid of words like "the" and "a" and "in" and "of", because they are not useful. I also want to only see the top words
top_count = df_count[(df_count['a'] != 'the') & (df_count['a'] != 'a') & (df_count['b']>5) & (df_count['a']!='of') & (df_count['a']!='it') & (df_count['a'] != 'to')& (df_count['a']!='in')& (df_count['a']!='on')& (df_count['a']!='is')]
top_count
```

```python
#Now I want to compare songs that do and don't make it to the top 40
all_names_top40 = ' '.join(list(billboard_clean['track'][billboard_clean.best_ranking <= 40])) 
all_names_notTop40 = ' '.join(list(billboard_clean['track'][billboard_clean.best_ranking > 40])) 

#following the same steps as before, but separating out songs that made it into the top 40 and ones that didn't
all_names_top40 = re.sub("[^A-Za-z0-9 ]", '', all_names_top40.lower()).split(' ')
all_names_notTop40 = re.sub("[^A-Za-z0-9 ]", '', all_names_notTop40.lower()).split(' ')

counter_items_top40 = Counter(all_names_top40)
counter_items_notTop40 = Counter(all_names_notTop40)

df_count_top40 = pd.DataFrame(counter_items_top40.items(), columns=['a', 'b'])
df_count_notTop40 = pd.DataFrame(counter_items_notTop40.items(), columns=['a', 'b'])

df_count_top40 = df_count_top40.sort_values('b', ascending=False)
df_count_notTop40 = df_count_notTop40.sort_values('b', ascending=False)

#Each will have different top songs, so I will do all for now, and limit the number included later
df_count_top40 = df_count_top40[(df_count_top40['a'] != 'the') & (df_count_top40['a'] != 'a')  & (df_count_top40['a']!='of') & (df_count_top40['a']!='it') & (df_count_top40['a'] != 'to')& (df_count_top40['a']!='in')& (df_count_top40['a']!='on')& (df_count_top40['a']!='is')]
df_count_notTop40 = df_count_notTop40[(df_count_notTop40['a'] != 'the') & (df_count_notTop40['a'] != 'a') & (df_count_notTop40['a']!='of') & (df_count_notTop40['a']!='it') & (df_count_notTop40['a'] != 'to')& (df_count_notTop40['a']!='in')& (df_count_notTop40['a']!='on')& (df_count_notTop40['a']!='is')]

#Merging them on the words for easier comparison
merged_top_words = pd.merge(df_count_top40, df_count_notTop40, how='outer', on='a', suffixes=['_top40', '_notTop40'])

#I want ton convert the numbers to percents for comparison (since each column has a different total)
merged_top_words['Top_40_percent'] = merged_top_words.b_top40*100/merged_top_words.b_top40.sum()
merged_top_words['Non_top_40_percent'] = merged_top_words.b_notTop40*100/merged_top_words.b_notTop40.sum()

#I want to show these words as a ratio so it's easier to show the differences
merged_top_words['ratio'] = (merged_top_words.Top_40_percent)/(merged_top_words.Non_top_40_percent)

#I only want to include words that were included in the previous visualization, since these are known to be top words
merged_top_words.index = merged_top_words.a
merged_top_words = merged_top_words.ix[list(top_count.a)]
merged_top_words['count of words in all'] = list(top_count.b)
merged_top_words
```

## Conclusions: 
I would recommend to my client the following. If you want to look at the Billboard Top 100 charts and figure out which newly-entered songs have a good chance of hitting the tops of the charts and lasting a long time:
- Take a song's first week's ranking with a grain of salt. Every subsequent week is more predictive of overall success, so keep track of their trajectories week-to-week.
- You're better off betting on songs that get categorized as "Rock" - whether or ot they really fit the category - because they tend to do better.
- You're also better of betting on personal (i.e. not third-person) songs regarding love. They seem to be common on the charts, and appear to do somewhat better as well.
- Ignore the length of the track - if a song has made it to the charts, the the length doesn't seem to impact its performance.
- Don't discount songs that are rising slowly on the charts - many of them will have good staying power! As long as they don't start going back down...

Another insight I would share with my client, if she doesn't already know this, is that the Billboard Top 100 Songs may not be the 100 most popular songs. Songs get kicked off the list after 20 weeks if they're not above a rank of 50, so don't discount songs that just fell off the list - they may still have some popularity left in them.
