## Step 2: Clean data

##### Do some rudimentary cleaning. Rename any columns that are poorly named, shorten any strings that are too long, and check for missing values (replace them and explain rationale if it makes sense).

```python
#Creating a new dataset that I will clean in-place
billboard_clean = billboard.copy()
```

```python
#Dropping useless variables: "year" is all the same, so it adds nothing; dropping all columns with all NaNs
billboard_clean.dropna(how='all', axis=1, inplace=True)
billboard_clean.drop('year', axis=1, inplace=True)
billboard_clean.head() #checking if correct
```

```python
#Renaming some variables
#first, rename 'artist.inverted" to be simpler and 'time' to 'length' to be clearer.
billboard_clean.rename(columns = {'artist.inverted' : 'artist', 'time' : 'length'}, inplace = True)

#next, getting rid of superfluous characters
billboard_clean.columns = [col.strip('x') for col in billboard_clean.columns]

#converting decimals to underscores to make dot-notation possible with all columns
billboard_clean.columns = [col.replace('.', '_') for col in billboard_clean.columns]

billboard_clean.head() #tests if it all worked
```

```python
# Converting 1st_week to float for consistency with the other Nth.week columns (NaN values are not 
# supported in an integer dtype) so I can do things with them together and do the same things to them.
billboard_clean.loc[:, '1st_week'] = billboard_clean.loc[:, '1st_week'].astype(float)
billboard_clean.dtypes['1st_week'] #testing if it worked
```

```python
#Now I am going to convert some datatypes to something more useful
#convert dates form objects to datetimes
billboard_clean['date_entered'] = pd.to_datetime(billboard_clean['date_entered'], format='%Y-%m-%d')
billboard_clean['date_peaked'] = pd.to_datetime(billboard_clean['date_peaked'], format='%Y-%m-%d')

#check if it appeared to do it right
print(billboard_clean['date_entered'].head(), billboard['date.entered'].head())
print(billboard_clean['date_peaked'].head(), billboard['date.peaked'].head())

#test that it is now the correct dtype
billboard_clean[['date_entered', 'date_peaked']].dtypes
```

```python
# 'length' column is listed as a object-dtype but would be more useful as a float.
# 1st made a function to convert a string in format 'min:sec' to a float in minutes
def minsec_to_float(string):
    min_sec = string.split(':')
    min_float = float(min_sec[0]) + float(min_sec[1])/60
    return min_float
# Test the function
minsec_to_float('3:50')
```

```python
#apply the function to 'length', return to new column in case of errors/bugs/fails
billboard_clean['length'] = billboard_clean['length'].apply(minsec_to_float)
#alternatively:
# billboard_clean['length'] = billboard_clean['length'].apply(lambda x: float(string.split(':')[0]) + float(string.split(':')[1])/60)

# #print the new and old columns side-by-side to see if it seems to have worked
pd.DataFrame([billboard_clean['length'], billboard['time']]).head()
```

```python
# investigate 'time_float'
print(billboard_clean['length'].describe())
```

```python
#Now I want to create some useful calculated variables: Weeks on Chart, Highest Ranking:
billboard_clean.ix[:, 'weeks_on_chart'] = billboard_clean.ix[:, '1st_week':'65th_week'].notnull().sum(axis=1)
billboard_clean.ix[:, 'best_ranking'] = billboard_clean.ix[:, '1st_week':'65th_week'].min(axis=1)
billboard_clean.head() #test if it worked
```

```python
#I have to first count the frequency of each word using collections.
from collections import Counter 
import re
all_names = ' '.join(list(billboard['track'])) #will create a string with all rows concatenated together by a space 
#needed 'list()' because it is a pandas series otherwise, which .join() won't work on
all_names_simplified = re.sub("[^A-Za-z0-9 ]", '', all_names.lower()) #lowercases everything and removes special characters so it will count accurately
all_names_list = all_names_simplified.split(' ') #will split everything into its separate words
counter_items = Counter(all_names_list) #creates a dictionary of words:wordcount
df_count = pd.DataFrame(counter_items.items(), columns=['a', 'b']) #converts to dataframe so I can sort (used .items() to get rid of error)
df_count = df_count.sort_values('b', ascending=False) #sorts ascending by word count ('b')

#i want to get rid of words like "the" and "a" and "in" and "of", because they are not useful. I also want to only see the top words
top_count = df_count[(df_count['a'] != 'the') & (df_count['a'] != 'a') & (df_count['b']>5) & (df_count['a']!='of') & (df_count['a']!='it') & (df_count['a'] != 'to')& (df_count['a']!='in')& (df_count['a']!='on')& (df_count['a']!='is')]
top_count
```

```python
#for my more numeric data, I will want to summarize my data, and I want to make it look clean.
from scipy import stats
def summarize(column):
    d = stats.describe(column)
    print "no. values: {}".format(d[0])
    print "min: {:.3}, max: {:.3}".format(d[1][0], d[1][1])
    print "mean: {:.3}".format(d[2])
    print "median: {:.3}".format(np.median(column))
    print "st dev: {:.3}".format(d[3]**.5)
    print "skew: {:.3}".format(d[4])
    print "kurt: {:.3}".format(d[5])
```
