## Scenario and Problem Statement

```python
import pandas as pd
```

```python
liq = pd.read_csv('assets/Iowa_Liquor_sales_sample_10pct.csv')
```

```python
liq.head()
```

```python
liq.columns = liq.columns.str.lower().str.replace(' ', '_').str.replace('(', "").str.replace(')', "")
liq.columns
```

```python
liq.dtypes
```

```python
liq.date = pd.to_datetime(liq.date)
liq.store_number = liq.store_number.astype(str)
liq.category = liq.category.astype(str)
liq.vendor_number = liq.vendor_number.astype(str)
liq.item_number = liq.item_number.astype(str)
liq.county_number = liq.county_number.astype(str)
liq['state_bottle_cost_usd'] = liq.state_bottle_cost.str.strip('$').astype(float)
liq['state_bottle_retail_usd'] = liq.state_bottle_retail.str.strip('$').astype(float)
liq['sale_usd'] = liq.sale_dollars.str.strip('$').astype(float)
liq.drop(['state_bottle_cost', 'state_bottle_retail', 'sale_dollars'], axis=1, inplace = True)
liq.dtypes
```

## Explore the data

Perform EDA and Viz, such as histograms of transaction totals, bottles sold, etc.

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
liq.describe(include='all')
```

```python
for col in liq.columns:
    if liq[col].dtype == 'int64' or liq[col].dtype == 'float64':
        plt.figure(figsize=(10,4))
        sns.distplot(liq[col])
        plt.show()
```

There are some major outliers making it very hard to visualize this dataset.

```python
from scipy import stats
```

```python
#eliminate outliers (>3 standard deviations from the mean)
for col in liq.columns:
    if liq[col].dtype == 'int64' or liq[col].dtype == 'float64':
        plt.figure(figsize=(10,4))
        sns.distplot(liq[col][abs(stats.zscore(liq[col]))<3])
        plt.show()
```

- bottle volumes seem to almost always fall into a few major sizes
- bottles sold seem to mostly be in 1, 2, 6, or 12
- volume sold in gallons and liters is the same - liters appears to be the more exact (since everything else is in metric, and the documentation indicates it's a formula dividing by 3785.411784, which is pretty exact but seems messier)
- state bottle cost and retail bottle cost seems like pretty much a shift. Will have to see if they always just use a particular formula
- total sales is VERY right-skewed.

```python
sns.pairplot(liq.ix[:,(liq.dtypes=='int64') | (liq.dtypes=='float64')])
```

This confirms that state and retail price are a perfect relationship with each other (colinear), and same with volume sold in gallons vs.  liters.
