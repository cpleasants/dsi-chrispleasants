## Risks and Assumptions:
The big assumption of this dataset is that the data are accurate and the patterns in this data are consistent with patterns in the whole population and is applicable outside this individual application of the Titanic (i.e. the patterns would, at least to some extent, hold true to another disaster). The dataset only includes 891 total passengers, out of the 2,224 who were aboard that day (https://en.wikipedia.org/wiki/RMS_Titanic). Since just over 2/3 (67%) of the passengers aboard the day it sank died, and 61% of those in this dataset died, there's a good chance this dataset isn't perfectly representative of the titanic as a whole. Furthermore, age data is missing for about 20% of the data and cabin data is missing for about 77% of the data. Therefore the usability of these may be limited. 

## Data Dictionary:

Column Name|Data Type|Description|Notes
-----------|---------|-----------|-----
survived|Boolean|Whether or not a passenger survived|
p_class|String|Passenger class (1st, 2nd, or 3rd class)| proxy for socio-economic status (SES): 1st ~ Upper; 2nd ~ Middle; 3rd ~ Lower
name|String|Passenger name|
sex|String|Passenger sex, male of female|
age|Float|Passenger age|
sib_spouse|Integer|Number of siblings and/or spouses aboard|includes brothers, sisters, step-brothers, step-sisters, and married spouses
parent_child|Integer|Number of parents and/or children aboard|includes mother, father, son, daughter, step-son, and step-daughters
fare|Float|Passenger fare (ticket cost)|
cabin|String|Cabin number(s)|
embarked|String|Port of embarkation| C = Cherbourg; Q = Queenstown; S = Southampton

information source: https://www.kaggle.com/c/titanic/data
