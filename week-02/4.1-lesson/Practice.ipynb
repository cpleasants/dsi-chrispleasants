```python
import numpy as np
import scipy as sp
from scipy import stats
```

```python
one_sample_data = [177.3, 182.7, 169.6, 176.3, 180.3, 179.4, 178.5, 177.2, 181.8, 176.5]
```

```python
one_sample = stats.ttest_1samp(one_sample_data, 175.3)
print "The t-statistic is %.3f and the p-value is %.3f." % one_sample
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.distplot(one_sample_data)
```

```python
sample_one = one_sample_data
sample_two = [180, 181, 160, 176, 182, 177, 174, 177, 182, 175]
```

```python
stats.ttest_ind(sample_one, sample_two)
```

```python
stats.ttest_rel(sample_one, sample_two)
```

```python
sns.distplot(sample_two, hist=False)
plt.show()
sns.distplot(sample_one, hist=False)
plt.show()
```

```python
sns.distplot([s2-s1 for s1, s2 in zip(sample_two, sample_one)])
```

```python
np.mean([s2-s1 for s1, s2 in zip(sample_two, sample_one)])
```
