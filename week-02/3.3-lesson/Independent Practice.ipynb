```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/suneel0101/lesson-plan/master/crunchbase_monthly_export.csv", na_values=' -   ')
df.head()
```

```python
df.columns = df.columns.str.strip(' ')
df.columns
```

```python
df.dropna(axis=1, how='all', inplace=True)
df.head()
```

```python
#make a dummy variable from the 'market' column
pd.get_dummies(df[' market ']).head()
```

```python
#Clean the funding_total_usd column (it's an 'object' and should be an integer)
df.dtypes
```

```python
df[' funding_total_usd '] = pd.to_numeric(df[' funding_total_usd '].str.replace(',', ''), errors='coerce')
```

```python
#Create some pivots
pd.pivot_table(df, index = ' market ')
```
